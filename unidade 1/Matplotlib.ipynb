```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 100, 11)
y = x ** 2
```

```python
plt.plot(x, y, 'b')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Título do gráfico')
plt.show()
```

```python
plt.subplot(1,2,1) # linhas, colunas, index
plt.plot(x, y, 'b')
plt.subplot(1,2,2)
plt.plot(y, x, 'r');
```

```python
fig, axes = plt.subplots(figsize=(12,3))

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Título');
```

```python
fig, axes = plt.subplots(figsize=(12,3))

axes.plot(x, y, 'r', label="normal")
axes.plot(x, y+1000, 'b', label="+1000")
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Título');
axes.legend()
```
